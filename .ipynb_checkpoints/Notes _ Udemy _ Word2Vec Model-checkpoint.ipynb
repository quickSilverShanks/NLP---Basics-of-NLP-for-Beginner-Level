```python
# A word2vec model should be made using a lot of data, like complete wikipedia. But we just want to check the concept.
import nltk
import urllib
import bs4 as bs
import re
```

```python
from gensim.models import Word2Vec
```

```python
# Gettings the data source
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Global_warming').read()

# Parsing the data/ creating BeautifulSoup object
soup = bs.BeautifulSoup(source,'lxml')

# Fetching the data
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```python
# Preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',text)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\W',' ',text)
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
# Preparing the dataset
sentences = nltk.sent_tokenize(text)

sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
# Training the Word2Vec model
model = Word2Vec(sentences, min_count=1)

words = model.wv.vocab
words
```

```python
vector = model.wv['global']
print(len(vector))
vector #the vector representing word 'global' in 100 dimensions
```

```python
similar = model.wv.most_similar('warming') #words similar to 'warming'
similar #removing stopwords should have removed impurities here.
```

#### Improving the above model

```python
# Fetching the data
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text

# Preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',text)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'[@#\$%&\*\(\)\<\>\?\'\":;\[\]-]',' ',text)
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)

# Preparing the dataset
sentences = nltk.sent_tokenize(text)
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in nltk.corpus.stopwords.words('english')]

# Training the Word2Vec model
model = Word2Vec(sentences, min_count=1)

words = model.wv.vocab
len(words)
```

```python
vector = model.wv['global']
print(len(vector))
vector #the vector representing word 'global' in 100 dimensions
```

```python
similar = model.wv.most_similar('warming') #words similar to 'warming'
similar #so the model has improved a lot now.
```

### Using Pre-Trained Model

```python
from gensim.models import KeyedVectors
```

```python
filename = 'GoogleNews_Vector\GoogleNews-vectors-negative300.bin'
# WARNING : don't try to load the model as its >3GB unless RAM is preferably >=8GB. In case of 8GB, screen freezes for a while and RAM stays loaded until session is active.
model = KeyedVectors.load_word2vec_format(filename,binary=True)
```

```python
model.wv.most_similar('king')
```

```python
# king - man + woman = queen
model.wv.most_similar(positive=['king','woman'],negative=['man'])
```
