```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
```

```python
train = pd.read_pickle('train_final.pkl')
test = pd.read_pickle('test_final.pkl')
print(train.shape, test.shape)
drop_features=['title', 'user_review', 'developer', 'publisher', 'tags', 'overview', 'lemmz_review', 'lemmz_tags', 'lemmz_overview']
train.drop(drop_features, axis=1, inplace=True)
test.drop(drop_features, axis=1, inplace=True)
print(train.shape, test.shape)
```

```python
x_train, x_val, y_train, y_val = train_test_split(train.drop(['user_suggestion'], axis=1).iloc[:,1:],
                                                  train['user_suggestion'], train_size=0.75)
train_x = x_train.to_numpy()
val_x = x_val.to_numpy()
train_y = keras.utils.to_categorical(y_train)
val_y = keras.utils.to_categorical(y_val)
```

```python
print(train_x.shape, val_x.shape, train_y.shape, val_y.shape)
```

```python
def plot_nnres(res0n):
    rows = len(res0n.history['loss'])
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.title('Cross-Entropy Loss')
    plt.plot(range(1,rows+1),res0n.history['loss'], label='Training')
    plt.plot(range(1,rows+1),res0n.history['val_loss'], label='Validation')
    plt.legend()
    plt.subplot(122)
    plt.title('Accuracy')
    plt.plot(range(1,rows+1),res0n.history['accuracy'], label='Training')
    plt.plot(range(1,rows+1),res0n.history['val_accuracy'], label='Validation')
    plt.legend()
```

```python
# Create Model
model01 = Sequential([
    Dense(250, activation='relu'),
    Dropout(0.20),
    Dense(250, activation='relu'),
    Dropout(0.15),
    Dense(100, activation='relu'),
    Dense(100, activation='relu'),
    Dense(2, activation='sigmoid')
])

# Compile Model
model01.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
res01 = model01.fit(x=train_x, y=train_y, batch_size=25, epochs=50, validation_data=(val_x,val_y), use_multiprocessing=True)
```

```python
plot_nnres(res01)
```

```python
# Generating Submissions
test_np = test.iloc[:,1:].to_numpy()
pred = model01.predict_classes(test_np)

submission01 = pd.DataFrame({'review_id' : list(test['review_id']),
                            'user_suggestion' : list(pred)})
submission01.to_csv('submission01.csv', index=False)
# Rank : 229/303submissions, My Score : 0.7983302895903991, Top100 : 0.8521453949
```
